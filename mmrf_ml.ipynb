```python
import requests

# Define the HGNC REST API URL for fetching immunoglobulin genes
hgnc_url = "https://rest.genenames.org/fetch/status/Approved"

# Define headers for the request
headers = {
    "Accept": "application/json"
}

# Send a GET request to the HGNC API
response = requests.get(hgnc_url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Initialize a list to hold Ensembl IDs
    ensembl_ids = []
    
    # Loop through the retrieved genes
    for gene in data['response']['docs']:
        # Check if the gene is an immunoglobulin gene (based on symbol or other criteria)
        if 'IG' in gene['symbol']:
            # Extract Ensembl ID if available
            if 'ensembl_gene_id' in gene:
                ensembl_ids.append(gene['ensembl_gene_id'])
    
    # Print the list of Ensembl IDs
    print("Ensembl IDs of immunoglobulin gene transcripts:")
    for ensembl_id in ensembl_ids:
        print(ensembl_id)
else:
    print("Failed to fetch data from HGNC API. Status code:", response.status_code)


```
